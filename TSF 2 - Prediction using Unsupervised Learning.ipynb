Task 2 - Prediction using Unsupervised Learning

In this task, I will predict the optimum number of clusters and represent it visually from the given 'iris' dataset

Using scikit learn, pandas, numpy and matplotlib 

```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

Importing the data from the source

```python
# Load the iris dataset

iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

```python
x = iris_df.iloc[:, [0, 1, 2, 3]].values #selecting iris dataset features into variable x

from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

From the above graph, the optimum number of clusters is where the elbow occurs. This is when the 'within cluster sum of squares' (WCSS) doesn't decrease significantly with every iteration. 

```python
# Applying kmeans to dataset
x = iris_df.iloc[:, [0,1,2,3]].values

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
print(y_kmeans)
```

```python
kmeans.cluster_centers_
```

```python
# Visualising the clusters - On the first 2 columns

plt.scatter(x[y_kmeans ==0, 0],x[y_kmeans==0, 1], s = 100, c ='red', label = 'Iris-setose')
plt.scatter(x[y_kmeans ==1, 0],x[y_kmeans==1, 1], s = 100, c ='blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans ==2, 0],x[y_kmeans==2, 1], s = 100, c ='green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
```

Conclusion

The K-Means clustering technique was implemented successfully with the help of elow method to predict the optimum number of clusters.
